```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

data = pd.read_csv("/kaggle/input/screentime-app-details-dataset/Screentime-App-Details.csv")
print(data.head())
```

```python
# check if the dataset has any null values or not

data.isnull().sum()
```

```python
print(data.describe())
```

```python
#lets first analyze the screen time of the user, look at the amount of usage of the apps

figure = px.bar(data_frame=data,
               x= "Date",
               y= "Usage",
               color= "App",
               title= "Usage")

figure.show()
```

```python
#number of notifications from the apps:

figure = px.bar(data_frame=data,
                x = "Date", 
                y = "Notifications", 
                color="App", 
                title="Notifications")

figure.show()
```

```python
#number of times the apps opened

figure = px.bar(data_frame=data, 
                x = "Date", 
                y = "Times opened", 
                color="App",
                title="Times Opened")
figure.show()
```

lets see the relationship between the number of notifications and the amount of usage:


```python
figure = px.scatter(data_frame = data, 
                    x="Notifications",
                    y="Usage", 
                    size="Notifications", 
                    trendline="ols", 
                    title = "Relationship Between Number of Notifications and Usage")
figure.show()
```
